```python
df=dfnames.sample(n=8,random_state=1)
df[['Name','Symbol']]
```

# 5

```python
symbols=df['Symbol'].values
```

```python
dfprices[symbols].plot(grid=True)
```

# 6

```python
stdmax=net_returns[symbols].std().to_frame('Std') 
stdmax[stdmax.Std==stdmax.Std.max()]
```

# 7

```python
gross_returns=dfprices[symbols]/dfprices[symbols].shift(1)
cum_gross_returns=gross_returns.cumprod()
cum_gross_returns.plot(grid=True)
```

```python
cum_gross_returns2=cum_gross_returns[-1:]
cum_gross_returns2.idxmax(1)
```

# 8

```python
log_gross_returns=np.log(gross_returns)
avg_log_gross_returns=log_gross_returns.mean()
avg_log_gross_returns=np.exp(avg_log_gross_returns)
avg_log_gross_returns
```
